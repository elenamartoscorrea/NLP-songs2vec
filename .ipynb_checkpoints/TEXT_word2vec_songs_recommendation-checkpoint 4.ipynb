The aim of this notebook is to make use of the word2vec model to find similar songs

```python
songs = pd.read_csv("hhgroups_merge_28_05.csv", header=0)
songs.head(2)
```

To train the word2vec model, we first need to build its vocabulary. To do that, I iterated over each song and added it to an array that can later be fed to the model.

* Se van a reemplazar todas los signos de puntuación y acentos para que no las cuente como palabras distintas.
Además, eliminamos la palabra estribillo y x2 ya que no son propias de la letra de la canción.

```python
import re
import string
```

```python
string.punctuation
```

```python
text_corpus = []
for song in songs['letra']:
    song = song.translate(str.maketrans('', '', string.punctuation))
    song = song.lower()
    song = re.sub('(¿|¡|estribillo|x2)','',song)
    words = song.split()
    text_corpus.append(words)

# Dimensionality of the resulting word vectors.
#more dimensions, more computationally expensive to train
#but also more accurate
#more dimensions = more generalized
num_features = 50
# Minimum word count threshold.
min_word_count = 1

# Number of threads to run in parallel.
#more workers, faster we train
num_workers = multiprocessing.cpu_count()

# Context window length.
context_size = 7

downsampling = 1e-1
```

* Después:

```python
song
```

```python
# Seed for the RNG, to make the results reproducible.
#random number generator
#deterministic, good for debugging
seed = 1

songs2vec = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features,
    min_count=min_word_count,
    window=context_size,
    sample=downsampling
)

songs2vec.build_vocab(text_corpus)
print(len(text_corpus))
```

```python
type(text_corpus)
```

```python
import time
start_time = time.time() 

songs2vec.train(text_corpus, total_examples=songs2vec.corpus_count, epochs=2)

if not os.path.exists("trained"):
    os.makedirs("trained")

songs2vec.save(os.path.join("trained", "songs2vectors.w2v"))

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
datos = data['The Project Gutenberg EBook of Don Quijote, by Miguel de Cervantes Saavedra']
```

```python
text_corpus2 = []
contar = 0
parrafos=[]
for line in datos:
    line = line.translate(str.maketrans('', '', string.punctuation))
    line = re.sub('(¿|¡|)','',line)
    words = line.lower().split()
    parrafos.extend(words)
    contar+=1
    if contar == 20:
        contar = 0
        parrafos = []
        text_corpus2.append(parrafos)
```

```python
# Minimum word count threshold.
min_word_count = 1

# Number of threads to run in parallel.
#more workers, faster we train
num_workers = multiprocessing.cpu_count()

# Context window length.
context_size = 7

downsampling = 1e-1
```

```python
print(text_corpus2[-1])
```

* Todas dentro del mismo contexto, porque separándolas por líneas los resultados no tenían mucho que ver.

```python
# Seed for the RNG, to make the results reproducible.
#random number generator
#deterministic, good for debugging
seed = 1
num_features = 50
window = 10 #10

text2vec = w2v.Word2Vec(
    sg=1,
    seed=seed,
    workers=num_workers,
    size=num_features, #Tamaño del vector denso para representar cada palabra
    min_count=min_word_count,
    window=context_size, #distancia máxima del contexto. Ventana más pequeña, elementos más relacionados
    sample=downsampling
)

text2vec.build_vocab(text_corpus2)
print(len(text_corpus2))
```

```python
type(text_corpus2)
```

```python
import time
start_time = time.time() 
# 5 número de epochs
text2vec.train(text_corpus2, total_examples=text2vec.corpus_count, epochs=5)

if not os.path.exists("trained"):
    os.makedirs("trained")

text2vec.save(os.path.join("trained", "text2vectors.w2v"))

print("--- %s seconds ---" % (time.time() - start_time))
```

* Calcular la suma de vectores normalizados de cada canción:

With the word vector embeddings in place, it is now time to calculate the normalised vector sum of each song. This process can take some time since it has to be done for each of 57,000 songs.

```python
songs = pd.read_csv("hhgroups_merge_28_05.csv", header=0)
```

```python
print(songs2vec['mujer'])
```

```python
def songVector(row):
    vector_sum = 0
    row = row.translate(str.maketrans('', '', string.punctuation))
    row = row.lower()
    row = re.sub('(¿|¡|estribillo|x2)','',row)
    words = row.split()
    for word in words:
        vector_sum = vector_sum + songs2vec[word]
    vector_sum = vector_sum.reshape(1,-1)
    normalised_vector_sum = sklearn.preprocessing.normalize(vector_sum)
    return normalised_vector_sum
```

```python
import time
start_time = time.time()
songs['song_vector'] = songs['letra'].apply(songVector)
```

**t-sne and random song selection** 

The songs have 100 dimensions each. Application of t-sne is memory intensive and hence it is slightly easier on the computer to use a random sample of the 57,000 songs.

```python
song_vectors = []
from sklearn.model_selection import train_test_split

train, test = train_test_split(songs, test_size = 0.9)


for song_vector in train['song_vector']:
    song_vectors.append(song_vector)

train.head(10)
```

I had a fairly measly 4gb machine and wasn't able to generate a more accurate model. However, one can play around with the number of iterations, learning rate and other factors to fit the model better. If you have too many dimensions (~300+), it might make sense to use PCA first and then t-sne.

```python
len(song_vectors)
```

```python
X = np.array(song_vectors).reshape((932, 100))

start_time = time.time()
tsne = sklearn.manifold.TSNE(n_components=2, n_iter=250, random_state=0, verbose=2)

all_word_vectors_matrix_2d = tsne.fit_transform(X)

print("--- %s seconds ---" % (time.time() - start_time))
```

```python
df=pd.DataFrame(all_word_vectors_matrix_2d,columns=['X','Y'])

df.head(10)

train.head()

df.reset_index(drop=True, inplace=True)
train.reset_index(drop=True, inplace=True)
```

Joining two dataframes to obtain each song's corresponding X,Y co-ordinate.

```python
two_dimensional_songs = pd.concat([train, df], axis=1)

two_dimensional_songs.head()
```
